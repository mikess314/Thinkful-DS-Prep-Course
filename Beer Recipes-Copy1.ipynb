```python
recipesByStyle = recipes.set_index("Style", drop = False)
#I want to slice out the outlier. But first I need to sort
```

```python
americanipa = recipesByStyle.loc["American IPA"]
americanipa = americanipa[americanipa["ABV"]<40]
americanipa = americanipa[americanipa["ABV"]>0.000]
#assumed a typographical error on extreme outliers of 49.96% and 45.16%
#assume no entry for 0.000 ABV entries
```

```python
americanipa['ABV'].describe()
```

```python
xmin = 0.000
xmax = 25.000
plt.hist(americanipa['ABV'], bins=25, range=[xmin,xmax], alpha=.8, rwidth=0.9)
plt.title("American IPA Alcohol by Volume")
plt.xlabel("ABV")
plt.ylabel("Frequency")
plt.show()
#increase size
```
